```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import shutil
from tqdm import tqdm
```

```python
# Set paths
IMAGES_PATH = 'C:/Users/LEGION/Desktop/extract text from table in image/data/images/'
ANNOTATIONS_PATH = 'C:/Users/LEGION/Desktop/extract text from table in image/data/anot/StructureLabelAddEmp/'

# List images
image_files = sorted(os.listdir(IMAGES_PATH))
annotation_files = sorted(os.listdir(ANNOTATIONS_PATH))

print(f"Found {len(image_files)} images")
print(f"Found {len(annotation_files)} annotation files")

# Example: Load 1 sample image + annotation
sample_image_file = image_files[0]
sample_annot_file = sample_image_file.replace('.png', '.txt')

# Read image
image = cv2.imread(os.path.join(IMAGES_PATH, sample_image_file))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Read annotation
with open(os.path.join(ANNOTATIONS_PATH, sample_annot_file), 'r', encoding='utf-8') as f:
    annotation = f.read()

# Show
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.title("Sample Document Image")
plt.axis('off')
plt.show()

# Print annotation
print(annotation)
```

```python
def create_mask_from_annotation(image_shape, annotation_file_path):
    mask = np.zeros(image_shape[:2], dtype=np.uint8)  # black mask

    with open(annotation_file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    for line in lines:
        try:
            if ';' not in line:
                continue  # skip malformed lines
            box_part, _ = line.strip().split(';')
            x1, y1, x2, y2 = map(int, box_part.split(','))
            cv2.rectangle(mask, (x1, y1), (x2, y2), color=255, thickness=-1)  # white filled box
        except:
            continue

    return mask
```

```python
# Generate mask for the sample
mask = create_mask_from_annotation(image.shape, os.path.join(ANNOTATIONS_PATH, sample_annot_file))

# Show the mask
plt.figure(figsize=(10, 10))
plt.imshow(mask, cmap='gray')
plt.title("Generated Mask from Annotation")
plt.axis('off')
plt.show()
```

```python
# Create dataset folders
OUTPUT_IMAGE_DIR = 'C:/Users/LEGION/Desktop/extract text from table in image/data/dataset/images'
OUTPUT_MASK_DIR = 'C:/Users/LEGION/Desktop/extract text from table in image/data/dataset/masks'

os.makedirs(OUTPUT_IMAGE_DIR, exist_ok=True)
os.makedirs(OUTPUT_MASK_DIR, exist_ok=True)

for image_file in tqdm(image_files):
    # Match annotation file
    annot_file = image_file.replace('.png', '.txt')
    image_path = os.path.join(IMAGES_PATH, image_file)
    annot_path = os.path.join(ANNOTATIONS_PATH, annot_file)

    if not os.path.exists(annot_path):
        continue  # Skip images with no annotations

    # Read image
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Create mask
    mask = create_mask_from_annotation(image.shape, annot_path)

    # Save image and mask
    cv2.imwrite(os.path.join(OUTPUT_IMAGE_DIR, image_file), cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
    cv2.imwrite(os.path.join(OUTPUT_MASK_DIR, image_file), mask)
```

```python
from tensorflow.keras.utils import Sequence
from sklearn.model_selection import train_test_split
import random

# Set paths
IMAGE_DIR = 'C:/Users/LEGION/Desktop/extract text from table in image/data/dataset/images'
MASK_DIR = 'C:/Users/LEGION/Desktop/extract text from table in image/data/dataset/masks'

IMG_SIZE = (256, 256)
BATCH_SIZE = 8
```

### Data Generator

```python
class TableSegmentationDataset(Sequence):
    def __init__(self, image_paths, mask_paths, batch_size, img_size, shuffle=True, augment=True):
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.batch_size = batch_size
        self.img_size = img_size
        self.shuffle = shuffle
        self.augment = augment
        self.on_epoch_end()

    def __len__(self):
        return int(np.ceil(len(self.image_paths) / self.batch_size))

    def __getitem__(self, idx):
        batch_images = self.image_paths[idx * self.batch_size:(idx + 1) * self.batch_size]
        batch_masks = self.mask_paths[idx * self.batch_size:(idx + 1) * self.batch_size]

        images = []
        masks = []

        for img_path, mask_path in zip(batch_images, batch_masks):
            img = cv2.imread(img_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = cv2.resize(img, self.img_size)

            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
            mask = cv2.resize(mask, self.img_size)
            mask = (mask > 127).astype(np.float32)
            mask = np.expand_dims(mask, axis=-1)

            if self.augment:
                if random.random() < 0.5:
                    img = cv2.flip(img, 1)  # horizontal flip
                    mask = cv2.flip(mask, 1)
                if random.random() < 0.5:
                    img = cv2.flip(img, 0)  # vertical flip
                    mask = cv2.flip(mask, 0)
                if random.random() < 0.5:
                    angle = random.choice([90, 180, 270])
                    img = self.rotate_image(img, angle)
                    mask = self.rotate_image(mask, angle)

            img = img / 255.0

            images.append(img)
            masks.append(mask)

        return np.array(images), np.array(masks)

    def on_epoch_end(self):
        if self.shuffle:
            combined = list(zip(self.image_paths, self.mask_paths))
            random.shuffle(combined)
            self.image_paths, self.mask_paths = zip(*combined)

    def rotate_image(self, image, angle):
        if angle == 90:
            return cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
        elif angle == 180:
            return cv2.rotate(image, cv2.ROTATE_180)
        elif angle == 270:
            return cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)
        return image
```

### Build The Model

```python
from tensorflow.keras import layers, models

def build_unet(input_shape):
    inputs = layers.Input(shape=input_shape)

    # Encoder
    c1 = layers.Conv2D(64, 3, activation='relu', padding='same')(inputs)
    c1 = layers.Conv2D(64, 3, activation='relu', padding='same')(c1)
    p1 = layers.MaxPooling2D()(c1)

    c2 = layers.Conv2D(128, 3, activation='relu', padding='same')(p1)
    c2 = layers.Conv2D(128, 3, activation='relu', padding='same')(c2)
    p2 = layers.MaxPooling2D()(c2)

    # Bottleneck
    b = layers.Conv2D(256, 3, activation='relu', padding='same')(p2)
    b = layers.Conv2D(256, 3, activation='relu', padding='same')(b)

    # Decoder
    u1 = layers.UpSampling2D()(b)
    u1 = layers.Concatenate()([u1, c2])
    c3 = layers.Conv2D(128, 3, activation='relu', padding='same')(u1)
    c3 = layers.Conv2D(128, 3, activation='relu', padding='same')(c3)

    u2 = layers.UpSampling2D()(c3)
    u2 = layers.Concatenate()([u2, c1])
    c4 = layers.Conv2D(64, 3, activation='relu', padding='same')(u2)
    c4 = layers.Conv2D(64, 3, activation='relu', padding='same')(c4)

    outputs = layers.Conv2D(1, 1, activation='sigmoid')(c4)

    return models.Model(inputs, outputs)
```
